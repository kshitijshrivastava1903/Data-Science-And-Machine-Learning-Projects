## K Means Cluster Creation

Now it is time to create the Cluster labels!

** Importing KMeans from SciKit Learn.**

```python
from sklearn.cluster import KMeans
```

** Creating an instance of a K Means model with 2 clusters.**

```python
km=KMeans(n_clusters=2)
```

**Fitting the model to all the data except for the Private label.**

```python
km.fit(college.drop('Private',axis=1))
```

** These are the cluster center vectors**

```python
km.cluster_centers_
```

## Evaluation

** Creating a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.**

```python
for i in range(777):
    college['Cluster']=0
for i in range(777):
    if college.iloc[i,0]=='Yes':
        college.iloc[i,18]=1
college        
```

```python
college['Cluster'].unique()
```

```python
l=km.labels_
l
```

** Creating a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.**

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,mean_absolute_error
print(confusion_matrix(college['Cluster'],l))
print(classification_report(college['Cluster'],l))
print(accuracy_score(college['Cluster'],l))
print(mean_absolute_error(college['Cluster'],l))
```
